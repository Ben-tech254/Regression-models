```python
#Loading and reading the dataset
housing_data = pd.read_csv('/content/Housing.csv')
```

```python
#One hot encoding
housing_data = pd.get_dummies(housing_data, columns = ['furnishingstatus'], drop_first= False)
```

```python
#Replacing yes and true values as 1 while no and false values as 0
housing_data = housing_data.replace({'yes' : 1, True : 1, False : 0, 'no' : 0})
```

##**Model creation**

```python
#Separating features from values.
array = housing_data.values

x = array[:, 1:15]

y = array[:, 0]

print(x.shape)

print(y.shape)
```

```python
#Splitting the dataset to features and the result
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)

print(x_train.shape)

print(x_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
#Importing the necessary regression models
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Appending all the models
models = []
models.append(('Decision Tree', DecisionTreeRegressor()))
models.append(('Support Vector', SVR()))
models.append(('Gradient Boosting', GradientBoostingRegressor()))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Linear Regression', LinearRegression()))
models.append(('K Neighbors', KNeighborsRegressor()))
```

```python
#Evaluation of the models
from sklearn.model_selection import cross_val_score, KFold

for name, model in models:
  kfold = KFold(n_splits = 10, random_state = 42, shuffle = True)

  cv_results = cross_val_score(model, x_train, y_train, cv = kfold, scoring = 'r2')

  print(name, 'Results : - ', cv_results.mean())
```

```python
#Choosing model then training it
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
#Predictions now
result = model.predict(x_test)

print('The model predicted : ', result)
print('\n\n')

#Import r2_score
print('\n\n')
print('The correct answer is : ', y_test)
```

```python
#Evaluating correctness of model predictions
from sklearn.metrics import r2_score

percent = r2_score(result, y_test)

print(percent*100, '%')
```

```python
#Fitting model 1
model_1 = RandomForestRegressor()

model_1.fit(x_train, y_train)
```

```python
#Prediction
result_1 = model_1.predict(x_test)

print('Random Forest predicted : - ', result_1)

print('\n\n')

print('The correct answers are : - ', y_test)

percent_1 = r2_score(result_1, y_test)

print('\n\n')

print(percent_1*100, '%')
```
